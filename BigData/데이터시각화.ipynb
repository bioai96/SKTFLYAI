```python
import newspaper
import matplotlib.pyplot as plt
from wordcloud import WordCloud

font_path ='./sample/NanumGothic.ttf'
wordcloud = WordCloud(font_path = font_path, background_color="white", width = 400, height=400)

link = "https://n.news.naver.com/mnews/article/014/0004867095?sid=105"
article = newspaper.Article(link, language='ko')
article.download()
article.parse()


wordcloud = wordcloud.generate(article.text)
fig = plt.figure(figsize=(6,6))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
fig.savefig("wordcloud.png")
```

```python
import os 
import pandas as pd
csv = os.path.join('./sample/anime_master.csv')
df = pd.read_csv(csv,index_col="anime_id")
df.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(df['members'], df['rating'], alpha=0.5)
plt.show()
```

```python
df.loc[df['members']>=800000,['name','members']]
```

```python
df.loc[(df['members']>=600000) & (df['rating']>=8.5), ['name','rating']]
```

```python
types = df['type'].unique()
types
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
# 카테고리 별로 분리 해서 볼수 있게 
for t in types :
    x = df.loc[df['type'] == t, 'members']
    y = df.loc[df['type'] == t, 'rating']
    ax.scatter(x,y,alpha = 0.5, label = t)
    
ax.set_title("배급 종별로 그룹화한 데이터 산포도 그래프")
ax.set_xlabel('Members')
ax.set_ylabel('Rating')
ax.legend(loc='lower right', fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt

if os.name == 'posix':
    plt.rc("font", family="AppleGothic")

else:
    plt.rc("font", family="Malgun Gothic")

# plt.rc('font', family='NanumGothic')
# plt.rc('axes',unicode_minus=False)
fig = plt.figure(figsize = (18, 3))
ax = fig.add_subplot(111)
wt = np.array(range(len(df)))
w = 0.1
for i in df.columns:
    ax.bar(wt, df[i], width=w, label=i)
    wt = wt + w
ax.set_xticks(np.array(range(len(df))))
ax.set_xticklabels(df.index, ha='left')
ax.set_ylabel('누적 멤버수')
ax.set_yscale('log') # 한 값이 너무 높을 때
ax.legend()
plt.show()
```

```python
x = np.arange(5)
np.random.seed(0)
y = np.random.rand(15).reshape((3,5))
y1,y2,y3 = y
y1b = np.array(y1)
y2b = y1b + np.array(y2)
y3b = y2b + np.array(y3)
#누적막대 그래프 그리기
fig = plt.figure(figsize=(10,3))
ax = fig.add_subplot(111)
ax.bar(x,y3b,label="y3") #역순으로 보여줘야함
ax.bar(x,y2b,label="y2")
ax.bar(x,y1b,label="y1")
ax.legend()

plt.show()
```
