```python
import os
import requests
from bs4 import BeautifulSoup as bs

url = 'https://comic.naver.com/webtoon/detail?titleId=641253&no=399&weekday=fri'

html = requests.get(url).text
soup = bs(html, 'html.parser')

result = soup.select('.wt_viewer img')

if not (os.path.isdir('./webtoon')):
    os.makedirs(os.path.join('./webtoon'))
i=1    
for tag in result :
    print(tag['src'])
    img_url = tag['src']
    headers = {'Referer':img_url}
    img_data = requests.get(tag['src'], headers = headers)
    save_img = "./webtoon/" + str(i).zfill(3) + img_url[-4:]
    print(save_img)
    with open(save_img, 'wb') as f :
        f.write(img_data.content)
    i += 1
```
