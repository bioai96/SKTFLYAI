```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
csv = os.path.join('./sample/도로교통공단_월별 교통사고 통계_20201231.csv')
df = pd.read_csv(csv,index_col=0)
df
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(221)
ax1 = fig.add_subplot(222)
ax.plot(df.index,df['사망자수'], label = ['DEATH'])
ax1.plot(df.index,df['사고건수'], label = ['ACCIDENT'])
ax.set_ylabel('EVENT')
ax.set_xlabel('MONTH')
ax.legend()
plt.show()
```
